```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from data_viz_helpers import *
```

```python
df = pd.DataFrame(columns=['x', 'y', 'end_x', 'end_y'])

for filename in ['EPL_2019_match_data.json', 'EPL_2020_match_data.json', 'EPL_2021_match_data.json']:
    # Read data from JSON file
    with open(filename, 'r') as f:
        data = json.loads(f.read())
    
    for item in data.items():
        for event in item[1]['matchCentreData']['events']:
            if event['type']['displayName'] == 'Pass' \
               and 'GOALASSIST' in str(event).upper() \
               and 'FROMCORNER' not in str(event['qualifiers']).upper():
                new_row = pd.Series(dtype='float')
                new_row['x'] = event['x']
                new_row['y'] = event['y']
                new_row['end_x'] = event['endX']
                new_row['end_y'] = event['endY']
                df = df.append(new_row, ignore_index=True)
    
    print(df.shape)
```

```python
# Scale data to fit on 130x90 field
x = df['x']*130/100
y = df['y']*90/100
end_x = df['end_x']*130/100
end_y = df['end_y']*90/100

# Make plot
fig, ax = plt.subplots(figsize=[16,12])

ax = plot_field(ax)

ax = sns.kdeplot(x, y, clip=[[0,130], [0,90]], shade=True, shade_lowest=False, cmap='jet', alpha=0.7)

for i in range(len(x)):
    ax.plot([x[i], end_x[i]], [y[i], end_y[i]], color='black', alpha=0.2)
ax.scatter(x, y, color='black', alpha=0.2)

ax.set_xlim([-1, 131])
ax.set_ylim([-1, 91])

plt.show()
```
