```python
import ScraperFC as sfc
```

```python
# Scrape 2018-19 EPL goal and shot creation from FBRef
try:
    scraper = sfc.FBRef()
    df = scraper.scrape_goal_shot_creation(2019, "EPL")
    scraper.close()
except:
    print("Scraping failed :(")
    scraper.close()
df # df is an array. First element is "for" stats. Second is "against" stats
```

```python
# Scrape 2019-20 Ligue 1 table from Understat
try:
    scraper = sfc.Understat()
    df = scraper.scrape_league_table(2020, "Ligue 1")
    scraper.close()
except:
    print("Scraping failed :(")
    scraper.close()
df
```
